# Production: Churn Prediction

```python
import pandas as pd
new_data = pd.read_csv("new_unseen_data.csv")
new_data.drop(["Exited"],axis=1, inplace=True)
save_data = new_data.copy()
new_data
```

```python
new_data.drop(["RowNumber","CustomerId","Surname"],axis=1, inplace=True)
new_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for column in new_data.columns:
    if new_data.dtypes[column] == object:
        new_data[column] = label_encoder.fit_transform(new_data[column])
new_data.head()
```

## Application du modele

```python
import joblib
model = joblib.load("decision_tree.modele")

predict = model.predict(new_data)
print(predict)
```

```python
predict = ["YES" if val == 1 else "NO" for val in predict]
save_data["Churn"] = predict
save_data
```

## Export Data

```python
save_data.to_csv("predicticed_customer.csv", sep=',')
```
